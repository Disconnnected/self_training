```python
# pyplot API vs OO API
# pyplot API -> quickly
# OO API -> advance
```

```python
# pyplot API
x = np.linspace(0,10,1000)
y = np.sin(x)
# plt.plot(x,y)
plt.plot(x,y, color = "red", linestyle= "dashed", label = "Sin(x)")
plt.plot(x, np.cos(x), color = "blue", linestyle= "dashed", label = "Cos(x)")
plt.legend()
plt.xlabel("Variable X")
plt.ylabel("Sin(x)")
plt.xlim([0,10])
plt.ylim([-2,2])
# plt.axis([xmin, xmax, ymin,ymax])
plt.axis("tight")
plt.axis("equal")
plt.title("sin(x) vs cos(x)")
```

```python
# OO API
x = np.linspace(0,10,1000)
y = np.sin(x)

# fig, ax = plt.subplots()
fig, ax = plt.subplots(figsize=[5,5])

ax.plot(x,y)

ax.set(
    title="OO API",
    xlabel="x",
    ylabel="sin(y)"
)
```

## Khám phá dữ liệu

```python
# Vẽ biểu đồ phân phối của điểm số môn Toán
plt.figure(figsize=(8, 6))
sns.histplot(df['Toán'], bins=10, kde=True)
plt.title('Phân phối điểm môn Toán')
plt.xlabel('Điểm')
plt.ylabel('Số học sinh')
```

>Biểu đồ phân phối (histogram) là một cách thể hiện sự phân bố của dữ liệu trong các khoảng giá trị khác nhau. Khi xem biểu đồ phân phối, chúng ta sẽ nhìn thấy cách dữ liệu được chia thành các khoảng (hay còn gọi là bin) và có bao nhiêu dữ liệu nằm trong mỗi khoảng đó. Biểu đồ phân phối giúp chúng ta có cái nhìn tổng quan về sự phân bố của dữ liệu và có thể phát hiện ra những đặc điểm quan trọng.
>
>Trong trường hợp của biểu đồ phân phối điểm môn Toán đã vẽ:
>
>- Trục X (Điểm): Đây là trục ngang của biểu đồ, thể hiện giá trị điểm trong môn Toán. Các khoảng giá trị điểm (bin) sẽ nằm trên trục này.
>
>- Trục Y (Số học sinh): Đây là trục đứng của biểu đồ, thể hiện số lượng học sinh nằm trong từng khoảng điểm tương ứng. Ví dụ, nếu thấy một thanh cột lớn tại khoảng điểm 6, điều này có nghĩa là có nhiều học sinh có điểm là 6.
>
>- Các thanh cột (bars): Các thanh cột trên biểu đồ thể hiện số lượng học sinh nằm trong từng khoảng điểm. Ví dụ, nếu có một thanh cột dày và cao ở khoảng điểm 5, điều này có nghĩa là có nhiều học sinh có điểm xấp xỉ 5.
>
>- Đường cong màu xanh (KDE - Kernel Density Estimation): Đây là đường cong mịn màng màu xanh nằm trên các thanh cột. Dòng cong này giúp bạn ước tính hình dáng của phân phối dữ liệu dưới dạng mật độ xác suất. Khi dòng cong nâng lên ở một vùng nào đó, điều này cho thấy nhiều dữ liệu tập trung ở vùng đó.
>
> Tóm lại, biểu đồ phân phối giúp chúng ta nhận biết sự phân bố của điểm số môn Toán, đặc biệt là xác định các khoảng điểm có số lượng học sinh nhiều nhất và những vùng mà điểm số tập trung.

```python
# Vẽ ma trận biểu đồ tương quan giữa các môn học
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Ma trận tương quan giữa các môn học')
plt.show()
```

>Ma trận biểu đồ tương quan (correlation heatmap) giữa các môn học giúp chúng ta hiểu về mối quan hệ tương quan giữa các cặp môn học. Tương quan đo lường mức độ liên quan giữa hai biến, trong trường hợp này, là điểm số giữa các môn học.
>
>Để hiểu cách đọc ma trận biểu đồ tương quan:
>
>- Ô vuông trên đường chéo chính: Đây là những ô vuông màu đỏ chứa số 1.0. Điều này thể hiện mối tương quan hoàn hảo giữa một biến với chính nó, vì chúng ta đang so sánh biến với chính nó.
>
>- Các ô vuông màu khác: Các ô vuông nằm ngoài đường chéo chính (đường chéo từ trái trên đến phải dưới) chứa các giá trị tương quan từ -1.0 đến 1.0. Màu sắc của ô vuông biểu thị mức độ tương quan: màu càng tối thể hiện tương quan càng mạnh (gần 1 hoặc -1), màu càng sáng thể hiện tương quan càng yếu (gần 0).
>
>- Chú thích về màu sắc: Bên cạnh biểu đồ, có thể có một chú thích màu sắc, thường được đặt ở bên cạnh ma trận. Nó cho biết mức độ tương quan tương ứng với màu sắc.
>
>- Số liệu bên trong ô vuông: Nếu thấy dấu số bên trong ô vuông, đó là giá trị tương quan số liệu thực. Số liệu này thể hiện mức độ tương quan chính xác giữa hai biến.
>
>Khi nhìn vào ma trận biểu đồ tương quan, có thể thấy mối quan hệ giữa các cặp môn học. Một giá trị tương quan gần với 1.0 có nghĩa là có mối tương quan dương mạnh (tăng cùng nhau), trong khi giá trị tương quan gần -1.0 thể hiện mối tương quan âm mạnh (khi một biến tăng thì biến còn lại giảm). Giá trị tương quan gần 0 thể hiện mối tương quan yếu hoặc không có mối quan hệ tương quan.

```python
# Tìm học sinh có điểm cao nhất và thấp nhất trong mỗi môn học
max_scores = df.idxmax()
min_scores = df.idxmin()

print("Học sinh có điểm cao nhất:")
print(max_scores)

print("\nHọc sinh có điểm thấp nhất:")
print(min_scores)
```

```python
# Tạo biểu đồ boxplot để phát hiện biến đổi bất thường
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Biểu đồ boxplot của các môn học')
plt.xticks(rotation=45)
plt.show()
```

>Biểu đồ boxplot là một công cụ mạnh để phát hiện sự biến đổi bất thường và hiểu về phân phối của dữ liệu trong mỗi biến
>
>- Hộp (Box): Phần hộp chứa giữa một phần tử thứ nhất và phần tử thứ ba của tập dữ liệu (25% và 75% phân vị), gọi là khoảng giữa tứ phân vị (IQR). Khoảng giữa tứ phân vị thể hiện phạm vi mà dữ liệu phân bố chủ yếu.
>
>- Đường kẻ ở giữa hộp (Median): Đây là giá trị trung vị của dữ liệu, tức là giá trị nằm ở giữa khi tập dữ liệu được sắp xếp theo thứ tự.
>
>- Vành (Whiskers): Vành trên và dưới biểu diễn khoảng từ giá trị thấp nhất/kéo đến phân vị thứ nhất (Q1 - 1.5IQR) đến giá trị cao nhất/kéo đến phân vị thứ ba (Q3 + 1.5IQR). Dữ liệu nằm ngoài khoảng này thường được coi là điểm ngoại lai (outliers).
>
>- Các điểm ngoại lai (Outliers): Những điểm nằm ngoài vùng vạch hình thang (vành) thường được hiển thị dưới dạng các điểm riêng biệt. Chúng có thể là dữ liệu biến đổi bất thường hoặc là các giá trị sai lệch so với phân phối chung.
>
>Biểu đồ boxplot giúp xác định sự biến đổi bất thường, phân phối dữ liệu và trung vị của biến, cũng như phân phối của dữ liệu xung quanh median. Nếu thấy rất nhiều điểm ngoại lai hoặc biểu đồ không đối xứng, điều này có thể gợi ý rằng dữ liệu của bạn có những biến đổi bất thường hoặc chưa tuân theo phân phối chuẩn.

```python
# - phân loại học sinh thành: "Ở lại lớp", "Trung bình", "Khá", "Giỏi", biết:
#     - học sinh "Giỏi"       : Nếu điểm trung bình >= 8.0, và không có môn nào dưới 6.5 điểm
#     - học sinh "Khá"        : Nếu điểm trung bình >= 6.5, và không có môn nào dưới 5 điểm
#     - học sinh "Trung bình" : Nếu điểm trung bình >= 5, và không có môn nào bị liệt
#     - "Ở lại lớp"           : Nếu học sinh đó dính bất kì một điểm liệt nào

# Tạo DataFrame chứa điểm liệt của từng học sinh
diem_liet_hs = df < 2.5

# Tính tổng số điểm từng học sinh
tong_diem_moi_hs = df.sum(axis=1)

# Tính điểm trung bình cho cả năm học của từng học sinh
diem_trung_binh_hs = (tong_diem_moi_hs / df.shape[1]).round(2)

# Khởi tạo danh sách để lưu phân loại của từng học sinh
phan_loai_hs = []

# Duyệt qua từng học sinh và xác định phân loại
for hs, diem_tb in diem_trung_binh_hs.items():
    if diem_tb >= 8.0 and not diem_liet_hs.loc[hs].any() and (diem_trung_binh_hs.loc[hs] >= 6.5).all():
        phan_loai_hs.append("Giỏi")
    elif diem_tb >= 6.5 and not diem_liet_hs.loc[hs].any() and (diem_trung_binh_hs.loc[hs] >= 5.0).all():
        phan_loai_hs.append("Khá")
    elif diem_tb >= 5.0 and not diem_liet_hs.loc[hs].any():
        phan_loai_hs.append("Trung bình")
    else:
        phan_loai_hs.append("Ở lại lớp")

# Thêm cột phân loại vào DataFrame
df['Phân loại'] = phan_loai_hs

# xoá cột này đi để còn chạy lại
# df = df.drop(columns=['Phân loại'])

# In ra kết quả phân loại của từng học sinh
print(df[['Phân loại']])

```

```python
# - Nâng cao: Các thí sinh có quyền lựa chọn khối thi cho bản thân. cho biết các khối thi:
#     - Khối A    : Toán - Lý - Hoá
#     - Khối A1   : Toán - Lý - Anh
#     - Khối B    : Toán - Hoá - Sinh
#     - Khối C    : Văn - Sử - Địa
#     - Khối D    : Văn - Toán - Anh

#     ==> hãy cho biết: 2 nguyện vọng (2 khối thi) mà thí sinh đạt điểm cao nhất
#     (Trường hợp 2 khối thi có số điểm bằng nhau thì lấy cả 2)
#     Ví dụ: thi sinh Trần tuấn thông:
#             - Khối A    : Toán - Lý - Hoá    = (5 + 4 + 8) = 17
#             - Khối A1   : Toán - Lý - Anh    = (5 + 4 + 9) = 18
#             - Khối B    : Toán - Hoá - Sinh  = (5 + 8 + 2) = 15
#             - Khối C    : Văn - Sử - Địa     = (7 + 3 + 7) = 17
#             - Khối D    : Văn - Toán - Anh   = (7 + 5 + 9) = 21

#             ==> 2 nguyện vọng mà Trần Tuấn Thông đạt điểm cao nhất là khối D với 21 điểm và khối A1 với 18 điểm

# Tính điểm cho từng khối thi cho mỗi học sinh
df['Khối A'] = df['Toán'] + df['Lý'] + df['Hoá']
df['Khối A1'] = df['Toán'] + df['Lý'] + df['Anh']
df['Khối B'] = df['Toán'] + df['Hoá'] + df['Sinh']
df['Khối C'] = df['Văn'] + df['Sử'] + df['Địa']
df['Khối D'] = df['Văn'] + df['Toán'] + df['Anh']

# Lựa chọn 2 khối thi có điểm cao nhất cho từng học sinh
df['NV 1'] = df[['Khối A', 'Khối A1', 'Khối B', 'Khối C', 'Khối D']].idxmax(axis=1)
df['Điểm NV 1'] = df[['Khối A', 'Khối A1', 'Khối B', 'Khối C', 'Khối D']].max(axis=1)
df['NV 2'] = df[['Khối A', 'Khối A1', 'Khối B', 'Khối C', 'Khối D']].apply(lambda row: row.nlargest(2).index.tolist(), axis=1)
df['Điểm NV 2'] = df[['Khối A', 'Khối A1', 'Khối B', 'Khối C', 'Khối D']].apply(lambda row: row.nlargest(2).values.tolist(), axis=1)

# In ra kết quả cho từng học sinh
print(df[['NV 1', 'Điểm NV 1', 'NV 2', 'Điểm NV 2']])
```

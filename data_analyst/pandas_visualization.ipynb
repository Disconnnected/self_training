```python
x = np.arange(-10,11)
```

```python
plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
```

```python
plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')
plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
```

```python
plt.figure(figsize=(12, 6), dpi=100, facecolor="white", frameon = False)
plt.plot(x, x ** 2)
"""
figsize(float, float): These parameter are the width, height in inches.
dpi : This parameter is the resolution of the figure.
num, facecolor,edgecolor,frameon,FigureClass,clear,linewidth,"""
```

```python
plt.plot(x, x ** 2, color = "red", linestyle= "dashed", label = "Sin(x)")
plt.legend()
plt.xlabel("Variable X")
plt.ylabel("Sin(x)")
plt.xlim([0,10])
plt.ylim([-2,2])
plt.axis("tight")
plt.axis("equal")
plt.title("sin(x) vs cos(x)")
```
